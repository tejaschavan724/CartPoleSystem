```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
m1 = 0.5 #Mass of cart(m1)

m2 = 0.2 #Mass of bob(m2)

l = 0.3 #length of pole(m)

g=10 #gravity

del_t = 0.01
time = np.arange(0,3,del_t) #time array

 
t = 0 #current time

theta_des = 3.14 #theta desired

x=0
x_dot=0
x_ddot = 0
#current theta,w,alpha
theta = 0
theta_dot = 0
theta_ddot = 0
#theta array
theta_arr = []
#reference array
r_arr = []
F_arr = []
x_arr = []
x_dot_arr = []
x_ddot_arr = []
#Parameters
kp = 100 #60
kd = 1000
ki = 1 #0

#0<theta<2pi
def convert_angle_to_0_2pi_interval(angle):
    new_angle = np.arctan2(np.sin(angle), np.cos(angle))
    if new_angle < 0:
        new_angle = abs(new_angle) + 2 * (np.pi - abs(new_angle))
    return new_angle

#error terms
old_e = 0
E = 0
#error array
e_arr = []
```

```python
#the pid controller
for i in range(300):
    x_dot_arr.append(x_dot)
    x_ddot_arr.append(x_ddot)
    x_arr.append(x)
    theta_arr.append(theta)
    r_arr.append(theta_des)
    e = theta_des - theta
    e_arr.append(e)
    e_dot = e - old_e
    E = E + e
    F = (kp*e) + (kd*e_dot) + (ki*E)
#     if F>60:
#         F = 60
#     elif F < -60:
#         F = -60
    F_arr.append(F)
    old_e = e
    print("F=",F)
    
    x_ddot = (F - m2*l*theta_ddot*math.cos(theta) + m2*l*(theta_dot**2)*math.sin(theta))/(m1+m2)
    theta_ddot = (-g*math.sin(theta)- x_ddot*math.cos(theta))/l
#Applying Newton's Laws    
    theta_dot = theta_dot + theta_ddot*del_t
    theta = theta + theta_dot*del_t + 0.5*theta_ddot*(del_t**2)
#     theta = convert_angle_to_0_2pi_interval(theta)
    
    x_dot = x_dot + x_ddot*del_t
    x = x + x_dot*del_t + 0.5*x_ddot*(del_t**2)
    print("theta =",theta)
    
    

print(theta_arr)
```

```python
plt.plot(theta_arr)
plt.plot(r_arr)
plt.plot()
plt.xlabel('Time')
plt.ylabel('theta')
plt.show()
```
